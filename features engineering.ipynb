```
def Hashtag(text):
  regex = "#(\w+)"
  hashtag_list = re.findall(regex, text)
  #print(len(hashtag_list))
  return len(hashtag_list)
```

```
def includeURL(text): 
  text=text.split()
  for txt in text:
    if re.match("((https?):((//)|(\\\\))+([\w\d:#@%/;$()~_?\+-=\\\.&](#!)?)*)",txt):
      return 1
  return 0 
```

```
def includephoneno(text): 
  text=text.split()
  for txt in text:
    if re.match('[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]',txt):
      return 1
  return 0 
```

```
s1 =[]

for i in df['tweet']:
  s1.append(Hashtag(i))
```

```
df['hashtag'] = s1
df
```

```
s2 =[]

for i in df['tweet']:
  s2.append(includeURL(i))

df['url'] = s2
df
```

```
s3 =[]

for i in df['tweet']:
  s3.append(includephoneno(i))

df['number'] = s3
df
```
